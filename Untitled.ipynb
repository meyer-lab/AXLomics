```python
import pandas as pd
from msresist.FileExporter import create_download_link

raw3 = pd.read_csv('./msresist/data/Raw/CombinedBR3_TR1&2_raw.csv')
raw3 = raw3[~raw3["Master Protein Descriptions"].str.contains("(blank)")]
ids = pd.read_csv('./msresist/data/Raw/prot_accessions_BR1&2.csv')
```

```python
ProtAcc = ids.iloc[:, 0]
ProtNames = ids.iloc[:, 1]
```

```python
NameToId = {}
for i in range(ids.shape[0]):
    NameToId[ProtNames[i]] = ProtAcc[i]
```

```python
raw3_acc = []
for i in range(raw3.shape[0]):
    raw3_acc.append(NameToId[raw3.iloc[i, 1]])
```

```python
raw3.insert(2, 'Master_Protein_Accessions', raw3_acc)
```

```python
accs = list(raw3.iloc[:, 2])
names = list(raw3.iloc[:, 1])
new_accs = [acc.split(";")[0] for acc in accs]
new_names = [name.split(";")[0] for name in names]

raw3.iloc[:, 2] = new_accs
raw3.iloc[:, 1] = new_names
```

```python
create_download_link(raw3, "CombinedBR3_TR1&2_raw.csv")
```
